```python
#Import Libraries
import glob
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.close('all') 
```

```python
#Retrieve all 4 datasets
dataset = sorted(glob.glob('data-*.csv'))
```

```python
#Define visualize and identify issue parameters
def visualize(d, title):
    data = d.plot.line(title=title)
    plt.show(data)
    identify_issue(d)
    return

def identify_issue(d):
    if d.loc[0]['max']==0 and d.loc[19]['max']==20:
        print("I doubt the integrity of this dataset.")
    elif d['min'].sum()==0:
        print("Why is this happening?")
    else:
        print("Dataset looks good.")
    return
```

```python
#Visualizations with Average, min and max of all datasets
for file in dataset:
    df = pd.read_csv(file, header=None)
    df_mean = df.mean(axis=0)
    df_max = df.max(axis=0)
    df_min = df.min(axis=0)
    d = pd.DataFrame({'Average':df_mean, 'max':df_max, 'min':df_min})
    visualize(d, file)
    
    #break
```
